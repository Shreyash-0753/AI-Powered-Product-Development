```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer   ##for feature extraction
from sklearn.metrics.pairwise import cosine_similarity        ## for similarity

# Load dataset
data = pd.read_csv("../admin1/Documents/tmdb_5000_credits.csv")

data = data[['title','cast','crew']]

# Handle missing values
data['cast'] = data['cast'].fillna('')
data['crew'] = data['crew'].fillna('')

# Combine features
data['tags'] = data['cast'] + " " + data['crew']

# Convert text to vectors
vectorizer = CountVectorizer(stop_words='english')
matrix = vectorizer.fit_transform(data['tags'])

# Cosine similarity matrix
similarity = cosine_similarity(matrix)

def recommend(movie):

    index = data[data['title'] == movie].index[0]

    scores = list(enumerate(similarity[index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print("Recommended Movies:\n")

    sim_scores = []

    for i in scores[1:6]:
        title = data.iloc[i[0]]['title']
        score = i[1] * 100
        sim_scores.append(score)

        print(title, "- Similarity:", round(score,2), "%")

    # Average similarity as accuracy
    accuracy = sum(sim_scores) / len(sim_scores)
    print("\nModel Accuracy (Average Cosine Similarity):", round(accuracy,2), "%")

# Test
recommend("Avatar")
```
